```python
# -*- coding: utf-8 -*-


from ckiptagger import data_utils, construct_dictionary, WS, POS, NER
import gc
import tensorflow as tf
import pandas as pd
# import keras

# 載入 ckiptagger 自訂詞典
User_Dict = {}
# with open("KCC Data/Dict/KccDict2020.txt","r", encoding='utf-8-sig') as UDicts:
with open("Ly Data/ly_dict.txt","r", encoding='utf-8-sig') as UDicts:
    for udic in UDicts:
        udWord = udic.strip().split(" ")
        if len(udWord) > 1:
            User_Dict[udWord[0]] = udWord[1]
        else:
            User_Dict[udWord[0]] = 150    # 未給定權重者一律賦予預設值 150                
dictionary = construct_dictionary(User_Dict)

# 載入模型
ws = WS("./data")
pos = POS("./data")
ner = NER("./data")

# 載入 StopWord
stopword_list = []
with open("KCC Data/Dict/KccStopWord2020.txt","r", encoding='utf-8-sig') as stopwords:
    for stopword in stopwords:
        stopword_list.append(stopword.strip())
        
        
# load transscript list
ly = pd.read_csv("Ly Data/ly_transcript.csv")
ly_lu = ly.loc[ly.委員姓名=='呂孫綾']
df_text = list(ly_lu["逐字稿"])

# 開始斷詞
# sentence_segmentation = True,  => To consider delimiters
# segment_delimiter_set = {",", "。", ":", "?", "!", ";"}), # This is the defualt set of delimiters
# recommend_dictionary = dictionary1, # words in this dictionary are encouraged
# coerce_dictionary = dictionary2, # words in this dictionary are forced
range(len(df_text))
```

```python
stayed_lst= []

strCut = ws( df_text, sentence_segmentation=True,
                segment_delimiter_set = {'"',"\r\n","，","「","」",",","？",'\n', "。", ":", "?", "!", ";", "、"},
                coerce_dictionary = dictionary)

# print(" / ".join(stayed_lst))
# print("")       
# print("------------CKIPTagger pos(str,....) ---------------------")
# pos_sentence_list = pos(strCut)
# print(pos_sentence_list)

# print("------------CKIPTagger ner(str,....) ---------------------")
# entity_sentence_list = ner(strCut, pos_sentence_list)

# print(entity_sentence_list)

print(strCut[0])
# del ws
# del WS
# del pos
# del POS
# del ner
# del NER
gc.collect()
```

```python
word_list = []
stayed_lst = []
print("------------CKIPTagger Word Segment ws(str,....) + 刪除標點及停用字 ---------------")
for l in range(len(strCut)):
    for word in strCut[l]:
        # 刪除停用字
        if word not in stopword_list:
            word_list.append(word)
    stayed_lst.append(word_list)
    word_list = []
```
